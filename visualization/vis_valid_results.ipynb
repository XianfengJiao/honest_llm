```python
import numpy as np
import pandas as pd
from glob import glob
```

```python
model_name = 'llama_7B'
dataset_name = 'tqa_mc2'
collect = [
    'pure',
    'all_100',
    'all_random_lower_bound000',
    'all_random_lower_bound050',
    'all_random_lower_bound075',
    'all_100_mean_alpha5',
    'all_random_lower_bound000_mean_alpha5',
    'all_random_lower_bound050_mean_alpha5',
    'all_random_lower_bound075_mean_alpha5',
    'all_100_pca_alpha5_n1',
    'all_100_pca_alpha5_n3',
    'all_random_lower_bound000_pca_alpha5_n1',
    'all_random_lower_bound050_pca_alpha5_n1',
    'all_random_lower_bound050_pca_alpha5_n3',
    'all_random_lower_bound075_pca_alpha5_n1',
    'all_100_pca_alpha5_n1_choose_by_pca',
    'all_random_lower_bound050_pca_alpha5_n3_choose_by_pca',
]
```

```python
answers_list = []
BLEURT_acc_list = []
MC1_list = []
MC2_list = []
bleu_acc_list = []
rouge1_acc_list = []
CE_Loss_list = []
KL_wrt_Orig_list = []
pre_path = f'/data/wtl/honest_llm/validation/{model_name}_{dataset_name}'
for clt in collect:
    print(clt)
    fold1_answers = pd.read_csv(glob(f'{pre_path}_{clt}/answer_{model_name}_seed_42*_fold_0.csv')[0])
    fold2_answers = pd.read_csv(glob(f'{pre_path}_{clt}/answer_{model_name}_seed_42*_fold_1.csv')[0])
    fold1_res = pd.read_csv(glob(f'{pre_path}_{clt}/summary_{model_name}_seed_42*_fold_0.csv')[0])
    fold2_res = pd.read_csv(glob(f'{pre_path}_{clt}/summary_{model_name}_seed_42*_fold_1.csv')[0])
    answers_list.append(fold1_answers[f'{model_name}'].values.tolist())
    MC1_list.append((fold1_res['MC1'][0] + fold2_res['MC1'][0]) / 2)
    MC2_list.append((fold1_res['MC2'][0] + fold2_res['MC2'][0]) / 2)
    BLEURT_acc_list.append((fold1_res['BLEURT acc'][0] + fold2_res['BLEURT acc'][0]) / 2)
    bleu_acc_list.append((fold1_res['bleu acc'][0] + fold2_res['bleu acc'][0]) / 2)
    rouge1_acc_list.append((fold1_res['rouge1 acc'][0] + fold2_res['rouge1 acc'][0]) / 2)
    CE_Loss_list.append((fold1_res['CE Loss'][0] + fold2_res['CE Loss'][0]) / 2)
    KL_wrt_Orig_list.append((fold1_res['KL wrt Orig'][0] + fold2_res['KL wrt Orig'][0]) / 2)
```

```python
df = pd.DataFrame({
        'collect':collect, 
        'MC1':MC1_list, 
        'MC2':MC2_list,
        'BLEURT acc': BLEURT_acc_list,
        'bleu acc':bleu_acc_list,
        'rouge1 acc':rouge1_acc_list,
        'CE Loss':CE_Loss_list,
        'KL wrt Orig':KL_wrt_Orig_list
    })
df
```
